```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

data = pd.read_csv('data.csv',sep=',')
display(data)
```

```python
features = ['ash','sodium','carb','cal']
x = data.loc[:,features].copy()
scaler = preprocessing.StandardScaler().fit(x)
data2 = scaler.transform(x)
print(data2)

target = data['brand'].copy()
print(target)
x_train, x_test, y_train, y_test = train_test_split(data2,target,test_size=0.3,random_state=0)
```

```python
LDA = LinearDiscriminantAnalysis(n_components=2)
data_projected = LDA.fit_transform(x_train,y_train)
print(data_projected)
```
