```python
import matplotlib.pyplot as plt

dataset = pd.read_csv(r"C:\Users\hanif\data-kategori-event-di-dki-jakarta-tahun-2019.csv")
total = dataset.groupby('jenis_event')['jumlah'].sum().reset_index()
total
```

```python
plt.figure(figsize=(10,5))
plt.bar(dataset["jenis_event"],dataset["jumlah"])
plt.xticks (rotation = 90)
plt.show()
```

```python
plt.hist(x = 'jumlah', data = dataset)
plt.xlabel('Jumlah ')
plt.ylabel('Frekuensi')
plt.title('Data kategori Event di Jakarta') 
num_bins = 10
plt.show()
```

```python
country = total['jenis_event']
rasio = total['jumlah']
fig = plt.figure()
ax = fig.add_axes([0,0,2,2]) 
ax.axis('equal')
ax.pie(rasio,labels = country,autopct = '%1.2f%%') 
plt.show()
```
